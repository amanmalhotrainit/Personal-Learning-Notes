```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#basic plotting
df = sns.load_dataset('tips')
df
```

```python
#Scatter plot
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Tip vs Bill")
```

```python
#line plot
sns.lineplot(x='size', y='total_bill', data=df)
plt.title("Size vs Bill")
```

```python
#cateogorical plot
#Bar plot
sns.barplot(x='day', y='total_bill', data=df)
plt.title("Day vs Bill")
```

```python
#box plot
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Day vs Bill")
```

```python
#voilin plot
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Day vs Bill")
```

```python
#histogram
sns.histplot(df['total_bill'], bins = 10, kde=True)
plt.title("Histogram")
```

```python
#KDE plot
sns.kdeplot(df['total_bill'], fill=True)
plt.title("KDE Plot")
```

```python
#pair plot
sns.pairplot(df)
```

```python
#heatmap
corr = df[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True)
```
