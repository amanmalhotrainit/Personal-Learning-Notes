```python
from google.colab import files

uploaded = files.upload()


import pandas as pd
import io

df = pd.read_csv(io.StringIO(uploaded['data.csv'].decode('utf-8')))
```

```python
df.head()
```

```python
#Handling missing values
#df.isnull() #fasle means no missing value but true means missing values
#this would give entire data
#we will use .any(axis) to get less but deep insights, this gives the row and coloumn which have a missing value
df.isnull().any(axis=1)
df.isnull().any(axis=0)
```

```python
df.isnull().sum() #gives number of missing value in each coloumn
```

```python
#df.fillna(value) the value which we will pass will replace the missing values
# repleacing with mean (new coloumn)
df["Sales_FillNA"] = df["Sales"].fillna(df["Sales"].mean())
df.head(10)
```

```python
#Renaming coloumns
df=df.rename(columns={'Date':'Sales_Date'}) #key value pair
df.head()
```

```python
#change datatypes
df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
#applying functions (creating a new coloumn with double value)
df['Value_changed'] = df['Value'].apply(lambda x: x*2)
df.head()
```

```python
#Data aggregating and grouping
#groupby(category)
group_mean = df.groupby('Product')['Value'].mean()
group_mean
```

```python
#grouping 2 categories
group_mean = df.groupby(['Product', 'Region'])['Value'].sum()
group_mean
```

```python
#aggregating functions
groupagg=df.groupby('Product')['Value'].agg(['sum','mean','count'])
groupagg
```

```python
#merging and joining dataframes
df1 = pd.DataFrame({
    'CustomerID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
})

# DataFrame with Order Information
df2 = pd.DataFrame({
    'OrderID': [1, 2, 3, 4, 5, 6],
    'CustomerID': [101, 102, 101, 104, 103, 106],
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer', 'Webcam'],
    'Amount': [1200, 25, 75, 300, 150, 45]
})
df1
```

```python
#merging on the key coloum
pd.merge(df1, df2, on='CustomerID', how='inner')
```

```python
pd.merge(df1, df2, on='CustomerID', how='outer')
```

```python
pd.merge(df1, df2, on='CustomerID', how='left') #df1 is given more priority
```

```python
pd.merge(df1, df2, on='CustomerID', how='right') #df2 is given more priority
```
