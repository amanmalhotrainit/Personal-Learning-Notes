```python
###upsampling and downsampling to deal with imbalanced data (900:100 in binary classification case)
```

```python
import numpy as np
import pandas as pd

#genrating random seed for reproducibilty
np.random.seed(123)

#creating dataframe with 2 classes
n_samples=1000
class_0_ratio=0.9
n_class_0=int(n_samples*class_0_ratio)
n_class_1=n_samples-n_class_0
```

```python
n_class_0,n_class_1
```

```python
# Create Class 0 data
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0 #900 0 in this dataframe
})

# Create Class 1 data
class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1 #100 1s in this dataframe
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
df.head()
```

```python
df['target'].value_counts()
```

```python
#upsampling
df_minority= df[df['target']==1]
df_majority= df[df['target']==0]
```

```python
from sklearn.utils import resample
#tries to create more points for minority to equalise with majority
```

```python
df_minority_upsampled= resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)
df_minority_upsampled
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
df_upsampled
```

```python
df_upsampled['target'].value_counts()
```

```python
#downsampling
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0 #900 0 in this dataframe
})

# Create Class 1 data
class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1 #100 1s in this dataframe
})
df=pd.concat([class_0,class_1]).reset_index(drop=True)
df.head()
df['target'].value_counts()
```

```python
df_minority= df[df['target']==1]
df_majority= df[df['target']==0]

df_majority_downsampled= resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42)
df_majority_downsampled
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])
df_downsampled
```

```python
df_downsampled['target'].value_counts() #eqialisation of data but dataloss
```
