```python
#1) Nominal/one hot encoding
# RED = [0,0,1], Green = [0,1,0], Blue = [1,0,0] ---> 3 categories(columns) created
#DISADV a) if 100 catoegires then 100 columns would be created
#       b) Sparse matrix leads to overfitting [1000 ice creams but only 100 eaters(one ice cream per person), 900 never get a chance to be tasted]
#hence avoid ohc for data with many catoegories
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
#create an instance of onehotencoder
encoder = OneHotEncoder()
```

```python
#fit and transform
encoded=encoder.fit_transform(df[['color']]).toarray()
encoded
```

```python
encoder_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoder_df
```

```python
#for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
#2) Label encoding ---> Red = 1, Blue = 2, Green = 3, unique numberical value for every category
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder() #object creation
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
#new value
lbl_encoder.transform([['blue']])
```

```python
#Disadv= Model may consider Red = 2 > Green = 1, model may take the encoding quite a bit literally
```

```python
#ordianal encoding = based on ranking eg highschool = 1, grad = 2, post grad = 3, phd = 4
# here the ranking would make sense
from sklearn.preprocessing import OrdinalEncoder
```

```python
df= pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
df
```

```python
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
#new data
encoder.transform([['small']])
```

```python
#Target guided ordinal coding
#used to import categorical variable based on relationship with target variable
#useful when we have cateogorical variable with large number of unique categories
```

```python
import pandas as pd
df = pd.DataFrame({
    'city':['new york','london','paris','tokyo','new york','paris'],
    'price':[200,150,300,250,150,320]
})
df
```

```python
#considering price as target variable and try to convert cateogirical feature (city) based on it
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
df
```
